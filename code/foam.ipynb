# CT scan acquisition simulation using [gVXR](https://gvirtualxray.sourceforge.io/) and CT reconstruction with [CIL](https://ccpi.ac.uk/cil/)

This example makes use of two open source libraries fro X-ray imaging. The first one is [gVXR](https://gvirtualxray.sourceforge.io/). It is used to simulate realistic radiographic images from a foam phantom. The second one is [CIL](https://ccpi.ac.uk/cil/). It implements many CT reconstruction algorithms, including the well-known FDK. The details of the CT scan acquisition are given in the table below. Both FDK and SIRT (iterative method) reconstructions were performed. 

![Image comparison of the FDK and SIRT reconstructions](../results/foam/comparison.png)

| Parameter | Value |
|-----------|-------|
| source-to-object distance (SOD) | 40 cm |
| object-to-detector distance (ODD) | 150 cm |
| source-to-detector distance (SDD) | 190 cm |
| detector resolution | 512 &times; 512 pixels |
| pixel pitch | 500 &times; 500 &mu;m |
| scintillator | 500 &mu;m of CsI|
| energy response of the detector | ![Plot of the energy response of the detector](../results/foam/detector-energy_response.png) |
| detector impulse response | ![Plot of the detector impulse response](../results/foam/detector-LSF.png) |
| tube anode material | tungsten |
| tube anode angle | 10&deg; |
| tube voltage | 160 kV |
| beam filtration | 1mm of Sn |
| tube exposure | 0.5 mAs |
| beam spectrum | ![Plot of the beam spectrum](../results/foam/beam-spectrum.png) |
| sample geometry | ![](../results/foam/wireframe.png) |
| sample material composition | Ti90Al6V4 |
| sample material density | 4.43 g/cm<sup>3</sup>|
| number of projection | 200 |
| first angle | 0&deg; |
| last angle | 360&deg; |
| number of flat images | 60 |

### View the created implicit surface (Interactively adjust foam mesh parameters)

The part below does not impact the final mesh rendering but is a simple way for the user to visualise what the isosurface threshold does in determining the final pore radius. By adjusting the isosurface threshold, the mask over the final image can be adjusted to see how much of the implicit surface will be rendered.

The isosurface threshold is a value that varies from the minimum to maximum density of the implicit surface such that:

* iso value = minimum density &rarr; None of the implicit surface is rendered as the mesh
* iso value = maximum density &rarr; All of the implicit surface is rendered as the mesh

A brightness/contrast slider is also added to show the foam pores more clearly

````python
# Write image stack to "output" directory
sitk.WriteImage(sitk_image, "../results/foam/imagestack.mha", useCompression=True);

# The minimum and maximum density values of the implicit surface
v_min = volume.min()
v_max = volume.max()

# Create sliders to scroll through the image stack, adjust brightness and change the isosurface threshold
z_slider = widgets.IntSlider(min = 0,
                             max = sitk_image.GetDepth()-1,
                             step = 1,
                             description = 'Z Slice')

brightness_slider = widgets.FloatRangeSlider(value = [v_min, v_max],
                                             min = v_min, max = v_max, 
                                             step = (v_max-v_min)*0.01,
                                             description = 'Brightness/Contrast:',
                                             readout_format = '.1f')

isosurface_value_slider = widgets.FloatSlider(min = v_min,
                                              max = v_max,
                                              step = (v_max-v_min)*0.01,
                                              description = 'isosurface threshold')
isosurface_value_slider.value = 0.36

# Function to display a slice of the image stack
def display_slice(z, brightness, isosurface_value):
    clear_output(wait=True)
    plt.figure(figsize=(10, 10))

    # Normalise the image
    image_min = brightness_slider.value[0]
    image_max = brightness_slider.value[1]

    normalised_image = (255 * (array_view[z_slider.value] - image_min) / (image_max - image_min))
    normalised_image[normalised_image < 0] = 0
    normalised_image[normalised_image > 255] = 255
    
    # Create a mask for isosurface threshold values
    mask = np.zeros(normalised_image.shape, dtype=np.uint8)
    mask[volume[z_slider.value] >= isosurface_value_slider.value] = 1

    # Display the normalised image along with mask
    plt.imshow(normalised_image.astype(np.uint8), cmap=plt.cm.Greys_r)
    plt.imshow(mask*255, 'jet', interpolation='none', alpha=0.5)

    plt.title(f"Slice {z} of the implicit function", fontsize=10)
    plt.axis("off")
    plt.show()

# Display the interactive widget
ui = widgets.HBox([z_slider, brightness_slider, isosurface_value_slider])

out = widgets.interactive_output(display_slice, {'z': z_slider, 
                                                 'brightness': brightness_slider, 
                                                 'isosurface_value': isosurface_value_slider})

display(ui, out)
````

### Creating a mesh from the above implicit function

The isosurface threshold for the mesh is set at 0.36 based on a value determined empirically. 

````python
implicit_surface_name = "foam_pores"

gvxr.removePolygonMeshesFromSceneGraph()

# Use implicit surface to create foam mesh
gvxr.createMeshFromImplicitSurface(implicit_surface_name, implicit_surface_name, isosurface_value_slider.value, "root", 1)
gvxr.createMeshFromImplicitSurface(implicit_surface_name, "cube", -0.01, isosurface_value_slider.value, "root", 1)

gvxr.setColor(implicit_surface_name, 0.0, 1.0, 0.0, 0.5)

# Create material for foam pores
gvxr.translateNode(implicit_surface_name, 0, 0, 0, "cm")
gvxr.setCompound(implicit_surface_name, "H2O")
gvxr.setDensity(implicit_surface_name, 1, "g/cm3")

# Create material for parent cube
gvxr.setMixture("cube", "Ti90Al6V4")
gvxr.setDensity("cube", 4.43, "g/cm3")

# Add the meshes to view in gVXR
gvxr.addPolygonMeshAsInnerSurface(implicit_surface_name)
gvxr.addPolygonMeshAsInnerSurface("cube")
````

## Simulate the CT acquisition and save the projections

````python
start = time.time();

# Simulate a CT scan acquisition
gvxr.computeCTAcquisition("../results/foam/projs", # Where to save the projections
                          "screenshots", # Where to save the screenshots
                          200, # Total number of projections
                          0, # First angle
                          True, # Include the last angle
                          360, # Last angle
                          60, # Number of flat images
                          0, 0, 0, "mm", # Centre of rotation
                          *gvxr.getDetectorUpVector()); # Rotation axis

stop = time.time();
runtime, unit = getRuntime(start, stop);
print("Execution time:", "{0:0.2f}".format(runtime), unit);
````

## Set the CT reconstruction parameters

````python
# Create the TIFF reader by passing the directory containing the files
reader = TIFFStackReader(file_name="../results/foam/projs", dtype=np.float32)

# Read in file, and return a numpy array containing the data
data_original = reader.read()

# The data is stored as a stack of detector images, we use the CILlabels for the axes
axis_labels = ['angle','vertical','horizontal']
````

````python
# Update the font size
font = {'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)
````

````python
# Use show2D to visualise a projection
show2D(data_original, origin="upper-left")
````

````python
# Interactive visualisation in Jupyter
islicer(data_original, origin="upper-left")
````

````python
# Normalisation
# Not strictly needed as the data was already corrected
data_normalised = data_original / data_original.max()

# Prevent log of a negative value
data_normalised[data_normalised<1e-9] = 1e-9

# Linearisation
data_absorption = -np.log(data_normalised)
````

````python
# Create the CIL geoemtry
geometry = AcquisitionGeometry.create_Cone3D(source_position=gvxr.getSourcePosition("cm"),
                                             detector_position=gvxr.getDetectorPosition("cm"),
                                             detector_direction_x=gvxr.getDetectorRightVector(),
                                             detector_direction_y=gvxr.getDetectorUpVector(),
                                             rotation_axis_position=gvxr.getCentreOfRotationPositionCT("cm"),
                                             rotation_axis_direction=gvxr.getRotationAxisCT())
                                         
# Set the angles, remembering to specify the units
geometry.set_angles(np.array(gvxr.getAngleSetCT()), angle_unit='degree')

# Set the detector shape and size
geometry.set_panel(gvxr.getDetectorNumberOfPixels(), gvxr.getDetectorPixelSpacing("cm"))

# Set the order of the data
geometry.set_labels(axis_labels)

# Set the angles, remembering to specify the units
geometry.set_angles(np.array(gvxr.getAngleSetCT()), angle_unit='degree')

# Set the detector shape and size
geometry.set_panel(gvxr.getDetectorNumberOfPixels(), gvxr.getDetectorPixelSpacing("cm"))
````

````python
# Shutdown the simulation engine
gvxr.terminate()
````

````python
# Display and save the geometry, does it look like a feasible CT scan set up?
fig = show_geometry(geometry);
fig.save("../results/foam/geometry.png");
fig.save("../results/foam/geometry.pdf");
````

````python
# Print details of the scanning geometry
print(geometry)
````

````python
# Prepare the data for the reconstruction
acquisition_data = AcquisitionData(data_absorption, deep_copy=False, geometry=geometry)
acquisition_data.reorder(order='tigre')
ig = acquisition_data.geometry.get_ImageGeometry()
````

````python
# Print details of the reconstructed volume
print(ig)
````

## Perform the CT reconstruction using the FDK algorithm and save the reconstructed volume

````python
# Perform the FDK reconstruction
fdk =  FDK(acquisition_data, ig)
recon = fdk.run()
````

````python
# Show a 2D slice
show2D(recon)
````

````python
# Interactive visualisation in Jupyter
islicer(recon)
````

````python
# Save the CT volume as a TIFF stack
TIFFWriter(data=recon, file_name=os.path.join("../results/foam/recons-FDK", "out")).write()
````

## Perform the CT reconstruction using the SIRT algorithm and save the reconstructed volume

````python
# Create projection operator using Astra-Toolbox.
acquisition_data.reorder(order='astra')
A = ProjectionOperator(ig, geometry, "gpu")
````

````python
# Create the initial guess
x0 = ig.allocate()

# non-zero constraint
constraint = IndicatorBox(lower=0)

# Instantiate the reconstruction algorithm
sirt = SIRT(initial=x0, operator=A, data=acquisition_data, constraint=constraint, max_iteration=500)
````

````python
# Perform 500 iterations
sirt.update_objective_interval = 50
sirt.run(500)

recon_sirt_noisy = sirt.solution
````

````python
# Show a 2D slice
show2D(recon_sirt_noisy)
````

````python
# Interactive visualisation in Jupyter
islicer(recon_sirt_noisy)
````

````python
# Show a comparison
fig = show2D([recon, recon_sirt_noisy, (recon-recon_sirt_noisy).abs()], \
       ['FBP', 'SIRT', 'difference'], \
       cmap="gray", num_cols=3, size=(15,15), origin='bottom-left', fix_range=True);

fig.save("../results/foam/comparison.png", dpi=300)
fig.save("../results/foam/comparison.pdf", dpi=600)
````

````python
# Save the CT volume as a TIFF stack
TIFFWriter(data=recon_sirt_noisy, file_name=os.path.join("../results/foam/recons-SIRT", "out")).write()
````
